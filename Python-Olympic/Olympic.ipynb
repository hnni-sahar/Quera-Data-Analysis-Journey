```python
top_10 = data[data['Medal'] == 'Gold']
top_10 = top_10['Country'].value_counts()
top10 = top_10.head(10).reset_index()
top10.columns = ['Country', 'Count']
```

```python
fig_top10, ax_top10 = plt.subplots(figsize=(10, 6))

# Create a bar chart
ax_top10.bar(top10['Country'], top10['Count'], color='#005f73')

# Set background color
fig_top10.patch.set_facecolor('#EAF4f4')
ax_top10.set_facecolor('#EAF4f4')

# Hide top, left, and right borders
ax_top10.spines['top'].set_visible(False)
ax_top10.spines['left'].set_visible(False)
ax_top10.spines['right'].set_visible(False)

# Set chart title
ax_top10.set_title('Top 10 Countries with the Most Gold Medals', fontsize=12, color='#22333b')

# Set y-axis label
ax_top10.set_ylabel('Number of Gold Medals', fontsize=10, color='#22333b')

# Set x-axis labels and rotation angle
ax_top10.set_xticks(range(len(top10['Country'])))
ax_top10.set_xticklabels(top10['Country'], color='#22333b')
ax_top10.tick_params(axis='x', labelrotation=45)

# Set tick color
ax_top10.tick_params(axis='x', colors='#22333b', length=0)
ax_top10.tick_params(axis='y', colors='#22333b', length=0)

# Save the chart
fig_top10.savefig('gold_medals_chart.png', bbox_inches='tight')

# Display the chart
plt.show()
```

```python
gender_count = data.groupby(['Year', 'Gender']).size().unstack(fill_value=0)
gender_count.columns.name = None
gender_count = gender_count.reset_index()
```

```python
fig_gen , ax_gen = plt.subplots(figsize=(10, 6))

# Set background color for the axes and the figure
plt.gca().set_facecolor('#EAF4f4')
fig_gen.set_facecolor('#EAF4f4')

# Filter data for women and men
women_data = gender_count[gender_count['Women'] == 'Women']
men_data = gender_count[gender_count['Men'] == 'Men']

# Plot data for women and men
plt.plot(gender_count['Year'], gender_count['Women'], label='Women', marker='o', color='#94d2bd')
plt.plot(gender_count['Year'], gender_count['Men'], label='Men', marker='o', color='#005f73')

# Fill the areas between the lines
plt.fill_between(gender_count['Year'], gender_count['Women'], gender_count['Men'], color='#005f73', alpha=0.6)
plt.fill_between(gender_count['Year'],0 , gender_count['Women'], color='#94d2bd', alpha=0.6)

# Add legend and set its background color
legend = plt.legend(loc='upper right')
legend.get_frame().set_facecolor('#EAF4f4') 

# Customize the axes
ax = plt.gca()
ax.spines['top'].set_visible(False)  # Remove the top spine
ax.spines['right'].set_visible(False)  # Remove the right spine
ax.spines['left'].set_visible(False)  # Remove the left spine

# Set x-axis limits
plt.xlim(1896, 2012)

# Set the title and label for the y-axis
plt.title('Male and Female Won Medals over the Years', fontsize=12, color='#22333b')
plt.ylabel('Number of Won Medals', fontsize=10, color='#22333b')

  
# Customize tick parameters
ax.tick_params(axis='x', length=0, labelcolor='#22333b')  # Hide x-axis ticks and set label color
ax.tick_params(axis='y', which='both', right=False, labelcolor='#22333b')  # Customize y-axis ticks
ax.yaxis.set_label_position('left')  # Place y-axis label on the left
ax.yaxis.tick_right()  # Show y-axis ticks on the right


# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
svk_data = data[data['Country'] == "SVK"]
grouped_data = svk_data.groupby(['Sport', 'Medal']).size().unstack(fill_value=0)
grouped_data.columns.name = None
grouped_data_svk = grouped_data.reset_index()
```

```python
fig_SVK, ax_SVK = plt.subplots(figsize=(10, 8))

plt.gca().set_facecolor('#EAF4f4')

outer_radius = 1
inner_radius = 0.7

outer_width = 0.3
inner_width = 0.3

# Define labels and colors
labels_outer = grouped_data_svk["Sport"]
labels_inner = ["Bronze", "Gold", "Silver"]
color_map = {'Gold': '#E9C46A', 'Silver': '#E0E1DD', 'Bronze': '#CD7F32'}
outer_colors = ['#d8f3dc', '#b7e4c7', '#95d5b2', '#74c69d', '#52b788', '#40916c']

# Create values for the inner pie chart
inner_vals = []
inner_colors = []
inner_labels = []

for _, row in grouped_data_svk.iterrows():
    for medal in labels_inner:
        medal_value = row[medal]
        if medal_value > 0:  # Include only non-zero values
            inner_vals.append(medal_value)
            inner_colors.append(color_map[medal])
            inner_labels.append(medal)

# Outer pie chart
patches1, texts1 = ax_SVK.pie(grouped_data_svk[["Bronze", "Gold", "Silver"]].sum(axis=1), 
                              radius=1, colors=outer_colors,
                              labels=labels_outer,
                              labeldistance=1.05,
                              wedgeprops=dict(width=outer_width, edgecolor="white"))

# Inner pie chart
patches2, texts2 = ax_SVK.pie(inner_vals, 
                              radius=1 - outer_width, colors=inner_colors,
                              labels=inner_labels,
                              labeldistance=0.75,
                              wedgeprops=dict(width=inner_width, edgecolor="white"))

plt.title('Count of Medals in each Sport', fontsize=12, color='#22333b')

plt.show()
```
